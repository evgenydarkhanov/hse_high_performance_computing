```python
import matplotlib.pyplot as plt
```

```python
N_1 = 10_000
result_1 = {
    1: 0.0492521,
    2: 0.0486151,
    4: 0.036494,
    8: 0.0319117,
    16: 0.0286991,
    24: 0.0264269 ,
}

x = [keys for keys, values in result_1.items()]
y = [values for keys, values in result_1.items()]

plt.plot(x, y)

plt.title(f"N = {N_1}")
plt.xlabel('num of processes')
plt.ylabel('time, seconds')
plt.grid()
```
